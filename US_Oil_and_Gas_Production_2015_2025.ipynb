```python
# 1. Import pustaka
import pandas as pd

# 2. Load dataset (ganti path sesuai lokasi file kamu)
df = pd.read_csv("/content/OGORBcsv_cleaned.csv")

# 3. Konversi kolom Production Date ke datetime
df['Production Date'] = pd.to_datetime(df['Production Date'], errors='coerce')

# 4. Cek hasil
print(df.info())
print(df.head())

```

```python
# Hapus koma dan ubah ke float
df['Volume'] = df['Volume'].replace({',': ''}, regex=True).astype(float)
```

```python
print(df['Commodity'].unique())
```

```python
# Agregasi per bulan dan jenis komoditas
monthly = df.groupby(['Production Date', 'Commodity'])['Volume'].sum().reset_index()

print(monthly.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
for commodity in monthly['Commodity'].unique():
    subset = monthly[monthly['Commodity'] == commodity]
    plt.plot(subset['Production Date'], subset['Volume'], label=commodity)

plt.title('Tren Produksi Minyak & Gas di AS (2015–2025)', fontsize=14)
plt.xlabel('Tahun')
plt.ylabel('Volume Produksi')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
